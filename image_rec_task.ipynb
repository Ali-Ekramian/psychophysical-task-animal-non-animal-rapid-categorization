```python
from psychopy import visual, core, event, clock
import numpy as np
import random
import os
import csv
```

### Functions

#### Main Function and Path

```python
def start_experiment():
    window = visual.Window(size=(1000, 800), fullscr=False, color='gray')
    noise = visual.ImageStim(window, image=r"/home/ali/Documents/VS_Code/Neuroscience/HW1/Pink_noise_cube.gif")
    animal_folder_path = r'/home/ali/Documents/VS_Code/Neuroscience/HW1/pics/A'
    non_animal_folder_path = r'/home/ali/Documents/VS_Code/Neuroscience/HW1/pics/NA'
    animal_files = os.listdir(animal_folder_path)
    non_animal_files = os.listdir(non_animal_folder_path)
    trials = 1200
    stimuli_order = animal_files + non_animal_files
    random.shuffle(stimuli_order)
    stimuli = load_stimuli(window, animal_files, non_animal_files, animal_folder_path, non_animal_folder_path, stimuli_order)
    stimulus_time = 0.020
    noise_time = 0.080
    fixation1_time = 0.50
    blank_time = 0.030
    fixation1 = visual.TextStim(window, text='+', color='white')
    fixation2 = visual.TextStim(window, text='+', color='black')
    blank = visual.TextStim(window, text='', color='black')
    ready_text = visual.TextStim(window, text='Y : Animal // N : NonAnimal', color='white')
    trial_data = []
    show_text_stimulus(window, ready_text)
    wait_for_key_press(["return"])
    rt_clock = clock.Clock()
    for trial in range(trials):
        present_fixation1(window, fixation1, fixation1_time)
        present_stimulus(window, stimuli[trial], stimulus_time)
        present_blank(window, blank, blank_time)
        present_noise(window, noise, noise_time)
        present_fixation2(window, fixation2)
        rt_clock.reset()
        response, reaction_time = collect_response()
        present_wait(window, blank)
        wait_for_key_press(["space"])
        res = calculate_correctness(response, stimuli_order, trial, animal_files, non_animal_files)
        trial_data.append([stimuli_order[trial], res, reaction_time])
    correctness_filepath = r"/home/ali/Documents/VS Code/Neuroscience/HW1/res1.csv"
    os.makedirs(os.path.dirname(correctness_filepath), exist_ok=True)
    save_correctness_file(correctness_filepath, trial_data)
    window.close()
    core.quit()
```

#### Stimulus and Timing Functions

```python
def load_stimuli(window, animal_files, non_animal_files, animal_folder_path, non_animal_folder_path, stimuli_order):
    stimuli = []
    for filename in stimuli_order:
        if filename in animal_files:
            stimulus_folder = animal_folder_path
        else:
            stimulus_folder = non_animal_folder_path
        stimulus_filepath = os.path.join(stimulus_folder, filename)
        stimulus = visual.ImageStim(window, image=stimulus_filepath)
        stimuli.append(stimulus)
    return stimuli
def show_text_stimulus(window, text_stimulus):
    text_stimulus.draw()
    window.flip()
def wait_for_key_press(key_list):
    event.waitKeys(keyList=key_list)
def present_stimulus(window, stimulus, duration):
    stimulus.draw()
    window.flip()
    core.wait(duration)
def present_fixation1(window, fixation, duration):
    fixation.draw()
    window.flip()
    core.wait(duration) 
def present_fixation2(window, fixation):
    fixation.draw()
    window.flip()
def present_blank(window, blank, duration):
    blank.draw()
    window.flip()
    core.wait(duration) 
def present_wait(window, blank):
    blank.draw()
    window.flip()
def present_noise(window, noise, duration):
    noise.draw()
    window.flip()
    core.wait(duration)
```

#### Saving Data Functions

```python
def collect_response():
    rt_clock = clock.Clock()
    response = event.waitKeys(keyList=['y', 'n'])[0]
    reaction_time = rt_clock.getTime()
    return response, reaction_time
def calculate_correctness(response, stimuli_order, trial, animal_files, non_animal_files):
    if (response == 'y' and stimuli_order[trial] in animal_files) or (response == 'n' and stimuli_order[trial] in non_animal_files):
        res = 1
    else:
        res = 0
    return res
def save_correctness_file(filepath, trial_data):
    with open(filepath, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Pic", "S4", "RA"])
        writer.writerows(trial_data)
```

### Start the Experinment

```python
start_experiment()
```
